```python
url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/refs/heads/main/TelecomX_Data.json'

import pandas as pd
import matplotlib.pyplot as plt
```

```python
clientes = pd.read_json(url)
```

#🔧 Transformação

```python
account_df = clientes['account'].apply(pd.Series)
clientes = pd.concat([clientes.drop(columns='account'), account_df], axis=1)
```

```python
clientes
```

```python
account_df = clientes['phone'].apply(pd.Series)
clientes = pd.concat([clientes.drop(columns='phone'), account_df], axis=1)
```

```python
account_df = clientes['customer'].apply(pd.Series)
clientes = pd.concat([clientes.drop(columns='customer'), account_df], axis=1)
```

```python
account_df = clientes['internet'].apply(pd.Series)
clientes = pd.concat([clientes.drop(columns='internet'), account_df], axis=1)

clientes
```

```python
account_df = clientes['Charges'].apply(pd.Series)
clientes = pd.concat([clientes.drop(columns='Charges'), account_df], axis=1)


clientes.head(10)
```

```python
clientes['Churn'].value_counts(normalize=True) # Mantém somente linhas válidas para análise de churn.
```

```python
cols_cat = [
    'gender', 'Contract', 'InternetService', 'PaymentMethod',
    'PhoneService', 'MultipleLines', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport',
    'StreamingTV', 'StreamingMovies', 'PaperlessBilling',
    'Partner', 'Dependents'
]

clientes[cols_cat] = clientes[cols_cat].apply(
    lambda col: col.astype(str).str.strip().str.title()
)
```

```python
clientes['tenure'].value_counts()
```

```python
bin_cols = [
    'PhoneService', 'MultipleLines', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport',
    'StreamingTV', 'StreamingMovies', 'PaperlessBilling',
    'Partner', 'Dependents'
]

clientes[bin_cols] = clientes[bin_cols].replace({'Yes': 1, 'No': 0})



# Padroniza para 0 e 1

# Facilita gráficos, médias e modelos

# Evita erro lógico em cálculos
```

```python
clientes.info()
```

```python
clientes['tenure'] = pd.to_numeric(clientes['tenure'], errors='coerce')
clientes['Monthly'] = pd.to_numeric(clientes['Monthly'], errors='coerce')
clientes['Total'] = pd.to_numeric(clientes['Total'], errors='coerce')


# Converte para número

# Valores inválidos viram NaN em vez de erro silencioso
```

```python
clientes[['tenure', 'Monthly', 'Total']].isna().sum()
```

```python
clientes = clientes.dropna(subset=['Total'])
```

```python
clientes = clientes[
    (clientes['tenure'] >= 0) &
    (clientes['Monthly'] > 0)
]
```

```python
clientes['churn_bin'] = clientes['Churn'].map({'Yes': 1, 'No': 0})
```

```python

clientes['Churn'].value_counts(normalize=True)
```

```python
clientes.head(10)
```

#📄Relatorio Final

As barras em vermelho representam os clientes com churn_bin = 1.

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 8))
axs = axs.ravel()

def plot_churn(ax, series, title, xlabel):
    valores = series.mul(100)
    cores = ['steelblue' if idx == 0 else 'red' for idx in valores.index]

    ax.bar(valores.index.astype(str), valores.values, color=cores)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel('Percentual (%)')
    ax.set_ylim(0, 100)
    ax.grid(True, axis='y', linestyle='--', alpha=0.6)

    for i, v in enumerate(valores.values):
        ax.text(i, v, f'{v:.1f}%', ha='center', va='bottom')

# 1️⃣ Contract
plot_churn(
    axs[0],
    clientes.groupby('Contract')['churn_bin'].mean(),
    'Churn por Tipo de Contrato',
    'Contrato'
)

# 2️⃣ Tenure (faixas)
tenure_bins = pd.cut(clientes['tenure'], bins=[0, 12, 24, 48, 72])
plot_churn(
    axs[1],
    clientes.groupby(tenure_bins)['churn_bin'].mean(),
    'Churn por Tempo de Contrato (Tenure)',
    'Meses de Contrato'
)

# 3️⃣ Partner
plot_churn(
    axs[2],
    clientes.groupby('Partner')['churn_bin'].mean(),
    'Churn por Parceiro',
    'Possui Parceiro'
)

# 4️⃣ Dependents
plot_churn(
    axs[3],
    clientes.groupby('Dependents')['churn_bin'].mean(),
    'Churn por Dependentes',
    'Possui Dependentes'
)

plt.suptitle('Análise de Churn — Fatores de Compromisso', fontsize=16)
plt.tight_layout()
plt.show()
```

Podemos ver que o churn por tipo de contrato mês a mês é maior doque os contratos de 1 ou 2 anos.

Em tenure, podemos ver que quanto mais tempo o cliente tem os serviços assinados, menor é a taxa de churn

Os contratantes que não tem parceiros ou dependentes também tem maior churn.

# Insights refinados — Linguagem de Negócios

## 📌 Tipo de contrato

Observa-se que clientes com contrato mensal apresentam uma taxa de churn significativamente maior quando comparados aos contratos de 1 ou 2 anos. Isso indica que a ausência de compromisso de longo prazo facilita a evasão, especialmente nos primeiros ciclos de uso do serviço.

Mensagem implícita para negócio:
Contratos mensais representam maior risco de churn e exigem ações específicas de retenção.

## 📌 Tempo de relacionamento (tenure)

A análise mostra uma relação inversa entre tempo de contrato e churn: quanto maior o tempo de permanência do cliente, menor a taxa de evasão. Isso sugere que clientes tendem a se tornar mais estáveis à medida que amadurecem no relacionamento com a empresa.

Mensagem implícita:
Os primeiros meses são críticos para retenção.

## 📌 Parceiros e dependentes

Clientes que não possuem parceiro ou dependentes apresentam maior propensão ao churn, o que pode estar associado a um perfil de menor estabilidade e maior flexibilidade para trocar de serviço.

Mensagem implícita:
Perfis com menor estabilidade pessoal demandam maior esforço de fidelização.

```python
servicos = [
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

clientes[servicos] = clientes[servicos].replace({'Yes': 1, 'No': 0})
```

```python
clientes['qtd_servicos'] = clientes[servicos].sum(axis=1)
```

```python
clientes.head(10)
```

```python
clientes['qtd_servicos'] = pd.to_numeric(
    clientes['qtd_servicos'],
    errors='coerce'
)
```

```python
clientes['qtd_servicos'].dtype
```

```python
clientes['qtd_servicos'].value_counts().sort_index()
```

```python
clientes['tenure_faixa'] = pd.to_numeric(clientes['tenure'], errors='coerce')
clientes = clientes.dropna(subset=['tenure'])
```

```python
clientes['tenure_faixa'] = pd.cut(
    clientes['tenure'],
    bins=[0, 12, 24, 48, 72],
    labels=['0–12', '13–24', '25–48', '49–72']
)
```

```python
import matplotlib.pyplot as plt

dados = (
    clientes
    .groupby(['tenure_faixa', 'qtd_servicos'])['churn_bin']
    .mean()
    .mul(100)
    .unstack()
)

fig, ax = plt.subplots(figsize=(12, 5))

dados.plot(
    ax=ax,
    marker='o',
    linewidth=2
)

ax.set_title(
    'Churn (%) por Tempo de Contrato e Quantidade de Serviços',
    fontsize=14
)
ax.set_xlabel('Tempo de Contrato (meses)')
ax.set_ylabel('Taxa de Churn (%)')

ax.set_ylim(0, 100)
ax.grid(True, linestyle='--', alpha=0.6)

ax.legend(
    title='Qtd. de Serviços',
    title_fontsize=10,
    fontsize=9
)

plt.tight_layout()
plt.show()
```

os clientes com 0 serviços tem chance de churn equivalente dos clientes com 1 a 4 serviços a mais nos primeiros anos. diminuindo apenas após longas datas com assinatura

pros clientes que tem muitos serivços (linha rosa 6) a taxa de churn zera no segundo ano. aumentando apenas após um longo período de assinatura.

mas, se olharmos pros clientes com assinatura de 5 serviços extras, vemos que esta sempre abaixo dos demais tipos de clientes, como um minimo para estabilização de churns.

# 📌 Serviços adicionais (valor percebido)

A presença de múltiplos serviços adicionais está associada a menores taxas de churn ao longo do tempo. Observa-se um limiar a partir do qual a adição de serviços contribui significativamente para a estabilização do cliente, especialmente após o segundo ano de contrato.

## 🎯 Objetivo

Analisar os principais fatores associados à evasão de clientes (churn), identificando padrões comportamentais e estruturais que impactam a retenção.

Normalização de variáveis categóricas

Análise exploratória baseada em taxas de churn (média de variável binária)

Segmentação por:

Tipo de contrato

Tempo de contrato

Perfil familiar

Quantidade de serviços contratados

Visualizações com escala percentual (0–100%) para facilitar interpretação

## 🔍 Principais Achados

Contratos mensais apresentam maior churn em comparação a contratos de longo prazo

A contratação de serviços adicionais reduz significativamente o churn ao longo do tempo, indicando maior valor percebido

# 📈 Conclusão

O churn não ocorre de forma aleatória, mas está fortemente associado a fatores de compromisso, tempo de relacionamento e percepção de valor. Estratégias de retenção devem priorizar os primeiros meses de contrato, incentivar contratos de maior duração e promover pacotes de serviços que aumentem o engajamento do cliente.
